##1. Anda diminta membangun klasifikator berbasis artificial neural networks (ANN) satu lapis untuk klasifikasi dua kelas ùë¶ ‚àà {0,1} dengan input ùëã = [ùë•1 ùë•2]. 

###a. Gambarkan ANN yang sesuai dan jelaskan masing-masing bagiannya. Jangan lupa tambahkan variabel bias x0 =1, sehingga ùëã = [ùë•0 ùë•1 ùë•2]. ANN akan digunakan untuk prediksi operasi AND, dimana hanya benar (y = 1) jika nilai kedua input adalah 1.

Berikut adalah deskripsi dan gambaran dari ANN yang sesuai untuk klasifikasi operasi AND:

Input Layer:
Pada input layer terdiri dari 3 node, node pertama layer ini mewakili bias dan memiliki nilai tetap x0 = 1.
Node kedua dan ketiga di layer ini mewakili input x1 dan x2.

Output Layer:
Node di layer ini mewakili output prediksi y.
Setiap node di layer ini terhubung dengan semua node di input layer.
Bobot yang digunakan untuk koneksi ini adalah W = [-30, 20, 20].

output : Dari ANN satu lapis Output yang akan dihasilkan berupa [0 atau 1] sesuai dengan logika AND

### Buktikan bahwa dengan menggunakan bobot W = [‚àí30 20 20] dan fungsi aktivasi sigmoid  

$ ùëî(ùëß) = \frac{1}{1+e^{-z}} $


###ANN tersebut dapat memprediksi operasi AND. 


```python
import numpy as np

# Fungsi Sigmoid
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Fungsi untuk prediksi
def prediksi(x, w):
    z = np.dot(x, w)
    y_hat = sigmoid(z)
    return np.round(y_hat)

# Definisikan input dan bobot
x = np.array([[1, 0, 0],  #Logika AND
              [1, 0, 1],
              [1, 1, 0],
              [1, 1, 1]])
w = np.array([-30, 20, 20])  # Bobot W


# Melakukan prediksi untuk setiap baris input
for i in range(len(x)):
    prediction = prediksi(x[i], w)
    print("Hasil prediksi untuk input", x[i, 1:], ":", int(prediction))
```

##2. Sekarang buatlah ANN dua lapis untuk memprediksi operasi XOR.

###Untuk itu ANN di rancang sbb.:
###‚Ä¢ lapisan input: 3 node (2 node input dan 1 node bias), 
###‚Ä¢ lapisan tersembunyi: 3 node (2 node sinyal masuk dan 1 node bias), dan 
###‚Ä¢ lapisan output: 1 node.
###‚Ä¢ Fungsi aktivasi kedua lapisan adalah sigmoid $ ùëî(ùëß) = \frac{1}{1+e^{-z}} $

###Selanjutnya jawab pertanyaan sbb.:
###a. Gambarkan ANN tersebut dan berikan penjelasan tiap bagian;

Cuplikan layar 2023-05-26 085108.png

Layer Input: Terdiri dari dua node input (x1 dan x2) yang mewakili nilai input untuk operasi XOR. Variabel bias x0 = 1 ditambahkan sebagai input pertama (x0) yang memiliki nilai tetap 1.


Lapisan Hide: Terdiri dari tiga node, dua di antaranya adalah node sinyal masuk (hidden node) dan satu adalah node bias. Hidden node menerima sinyal dari node input melalui koneksi dengan bobot yang sesuai. Variabel bias x0 = 1 juga ditambahkan sebagai input pertama pada node bias. Node-node ini akan melakukan transformasi non-linear pada sinyal masukan menggunakan fungsi aktivasi sigmoid.


Lapisan Output: Terdiri dari satu node output yang menerima sinyal dari node-node di lapisan tersembunyi melalui koneksi dengan bobot yang sesuai. 

output : akan berkerja sesuai logika XOR yang menghasilkan nilai 1/0









###b. Gunakan parameter bobot untuk kedua lapisan:

```python
import numpy as np

# Fungsi aktivasi sigmoid
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Fungsi Prediksi
def prediksi(x, w1, w2):
    h_input = np.dot(x, w1.T)  # Perhitungan input lapisan tersembunyi
    h_output = sigmoid(h_input)  # Aktivasi lapisan tersembunyi

    # Tambahkan bias ke lapisan tersembunyi
    h_output_with_bias = np.insert(h_output, 0, 1)

    o_input = np.dot(h_output_with_bias, w2.T)  # Perhitungan input lapisan output
    o_output = sigmoid(o_input)  # Aktivasi lapisan output
    
    return np.round(o_output)

    

# Definisikan input XOR
x = np.array([[1, 0, 0],  # Input X = [x0, x1, x2]
              [1, 0, 1],
              [1, 1, 0],
              [1, 1, 1]])

# Definisikan bobot untuk kedua lapisan
w1 = np.array([[-30, 20, 20],  # Matriks bobot untuk lapisan tersembunyi
               [10, -20, -20]])
w2 = np.array([[-30, 20, 20],
               [10,-20,-20]])  # Matriks bobot untuk lapisan output

for i in range(len(x)):
    prediction = prediksi(x[i], w1, w2)
    print("Hasil prediksi untuk input", x[i, 1:], ":", int(prediction[1]))


```
